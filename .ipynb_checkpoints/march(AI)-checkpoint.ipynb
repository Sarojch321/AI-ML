### Decision Trees
- Uses CART (Classificaton and Regression Techniques) => find out Gini index (measure of impurity or uncertainity)
- probability of miss classification

```python
# for color (Red Yellow)
## Red : 5
    # yes : 3
    # no : 2
gini_red = 1 - ((3/5)**2 + (2/5)**2)
print(gini_red)
```

```python
# for color (Red Yellow)
## yellow : 6
    # yes : 3
    # no : 6
gini_yellow = 1 - ((3/6)**2 + (3/6)**2)
print(gini_yellow)
```

```python
# gini_color = prob(red) * gini(red) + prob(yellow) * gini(yellow)
gini_color = 5/11 * gini_red + 6/11 * gini_yellow
print(gini_color)
```

```python
# for type (casual formal)
## formal : 4
    # yes : 1
    # no : 3
## casual : 7
    # yes : 5
    # no : 2
gini_type = 4/11 * (1- ((1/4)**2 + (3/4)**2)) + 7/11 * (1 - ((5/7)**2 + (2/7)**2))
print(gini_type)
```

```python
# from above both gini jasko low gini hunxa tyo root node hunxa
```

### Entropy - measures uncertainity

- Overfitting => performs very well in training data, but low in testing data
- Underfitting => A model that performs poorly on both the training and test data

```python
decision_tree = DecisionTreeClassifier(criterion = 'entropy')
decision_tree.fit(xtrain, ytrain)
```

```python
plt.figure(figsize = (12,12))
plot_tree(decision_tree, feature_names = columns, class_names = class_ , filled = True)
plt.show()
```

```python
ypred_train = decision_tree.predict(xtrain)
```

```python
ypred = decision_tree.predict(xtest)
```

```python
accuracy = accuracy_score(ytest, ypred) * 100
accuracy
```

```python
accuracy1 = accuracy_score(ytrain, ypred_train) * 100
accuracy1
```

- Accuracy for training > Accuracy for testing so this is overfitting
- A model that performs poorly on both the training and test data => underfitting

#### Assumption
- Stationary

  
##### ARIMA -> Autoregressive Integretion Moving Average

#### To check Stationary
- Plot
- Test (Augmented Dickey Fuller Test) Hypothesis

```python
from statsmodels.tsa.stattools import adfuller
```

```python
test = adfuller(dfn['Gold_Price'])
```

```python
test
```

```python
test_statistic = test[0]     #  calculated value
print(test_statistic)
p_value = test[1]
print(p_value)
```

- p-value < 0.05  - stationary
- p-value > 0.05  - not stationary

#### First make stationary
Differencing

```python
dfn['shift1'] = dfn['Gold_Price'].shift(1)
dfn.head()
```

```python
dfn['Diff'] = dfn['Gold_Price'] - dfn['shift1']
dfn.head()
```

```python
dfn.dropna(inplace = True)
dfn.head()
```

```python
testagain = adfuller(dfn['Diff'])
testagain[1]
```

```python
dfn['Diff'].plot()
plt.show()
```

```python
d = 1  # since after differencing by 1, data converted into stationary
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plot_pacf(dfn['Diff'], lags = 20)
plt.show()
```

```python
# first near to 0  se aagek lena
p = 1
```

```python
plot_acf(dfn['Diff'], lags = 20)
plt.show()
```

```python
q = 1
```

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(dfn['Gold_Price'], order = (1,1,1))  
# OR  ->  model = ARIMA(dfn['Diff'], order = (1,0,1))# order = (p, d, q)
arima_model = model.fit()
```

```python
result = arima_model.get_forecast(steps = 10)
forecasted = result.predicted_mean
forecasted
```
